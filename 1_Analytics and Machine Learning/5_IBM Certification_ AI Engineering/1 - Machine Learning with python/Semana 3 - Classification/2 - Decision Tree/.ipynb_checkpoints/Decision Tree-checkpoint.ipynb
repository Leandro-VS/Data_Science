# Decision Tree

__Decision Tree__ é um modelo de classificação baseado na estrutura de dados de arvore de decisão. Este modelo dividi o conjunto de treino em diferentes nodos, classificando as amostras em diferentes grupos, a divisão continua -de modo recursivo- até que a arvore tenha conjuntos homegêneos nos dois lados do seu ramo.<br>
Para decidir qual feature melhor dividirá o conjunto de dados, de modo a tornar os conjuntos gerados mais homogêneos possível, o modelo calculo o grau de Entropia de cada feature e depois o ganho de informação que terá após a divisão dela, ou seja, a entropia de cada conjunto após a divisão, é calculada a diferença entre essas duas medidas e assim o modelo determina qual feature melhor divide os dados. Este processo se repete até uma profundidade de arvore pré determinada ou algumas outras metricas de parada que podem ser inseridas no modelo

__Entropia:__
\begin{equation*}
    Entropy = -p(A)log(p(A)) - p(B)log(p(B))
\end{equation*}
<br><br>
__Ganho de Informação:__
\begin{equation*}
   \text{Info. Ganho} = \text {(Entropia Inicial)} - \text{(Peso da entropia após a divisão)} 
\end{equation*}


```python
#Libraries
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

## Sobre o Dataset

O dataset consiste de dados sobre pacientes que sofrem de uma mesma doença. Durante o tratamento cada paciente respondeu de forma positiva a um dentre 5 medicamentes testados: Drug A, Drug B, Drug c, Drug x and y.<br>
O trabalho aqui consiste em criar um modelo que encontre o medicamento correto para um futuro paciente com essa mesma doença. <br>
No dataset consta informações como Sexo, idade, Pressão sanguinea, dentre outras, de cada paciente e o medicamento com uma resposta positiva.

# Preparando os dados

```python
#Split
y = df['Drug']
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
X[:5]
```

Algumas features são categóricas e para o modelo devemos ter apenas features numéricas.

Outra maneira de fazer essa transformação é utilizando um pacote da lib Scikit-Learn, chamado Label Enconder

```python
from sklearn import preprocessing
```

```python
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F', 'M'])
X[:,1] = le_sex.transform(X[:, 1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit(['NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3])
```

```python
X[:5]
```

```python
y[:5]
```

# Modelo

Utilizaremos o método de train_test_split para generalizar nosso modelo e também para aplicar o processo de validação cruzada.

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=.3, random_state=3)
```

```python
X_trainset.shape, X_testset.shape, y_trainset.shape, y_testset.shape
```

```python
#Instanciando o modelo
#Criteria: Entropy, nos mostrará o Ganho de Informação de cada Nodo
drugtree = DecisionTreeClassifier(criterion='entropy', max_depth=4)
drugtree
```

## Treinando o modelo

```python
drugtree.fit(X_trainset, y_trainset)
```

## Previsões

```python
predTree = drugtree.predict(X_testset)
print(predTree)
```

# Evaluation

```python
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
print("DecisionTree's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```

```python
#Accuracy score sem sklearn
valid = (predTree == y_testset)
valid.mean()
```
